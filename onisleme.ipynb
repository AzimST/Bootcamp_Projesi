```python
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# İmage List
image_list=[]
```

```python
img_size = 64
```

```python
# dosyadan resimleri çekip işleme yapıyoruz ardından listeye atıyoruz
for i in range(0,10):
    file = glob.glob(f"/content/drive/MyDrive/spectrograms/{i}/*.png")
    # dosyadaki resim sayısını ekrana bastırıyoruz
    for spectrogram in file:
        try:
            # resim okuma
            img = cv2.imread(spectrogram,cv2.IMREAD_GRAYSCALE)
            # resize 
            resized_img = cv2.resize(img,(img_size,img_size))
            # grayscale
            # normalizasyon
            image_list.append([resized_img,i])
        except Exception as e:
            pass
            
```

```python
# label list
label_list=[]
image_raw_list=[]
```

```python
# image_listi label list ve image_raw_list ikiye ayırıyoruz
for data in image_list:
    label = data[1]
    img= data[0]
    label_list.append(label)
    image_raw_list.append(img)
```

```python
print(image_raw_list[800])
```

```python
# image listesini reshape yapıyoruz
image_raw_list = np.array(image_raw_list).reshape(-1,img_size,img_size,1)
```

```python
print(image_raw_list[800])
```

```python
x_train,x_temporary,y_train,y_temporary = train_test_split(image_raw_list,label_list,train_size=0.8)
```

```python
x_test,x_val,y_test,y_val = train_test_split(x_temporary,y_temporary,train_size=0.5)
```

```python
# dataseti kaydetme
np.save("/content/mobel_Data/X_train.npy",x_train)
np.save("/content/mobel_Data/X_test.npy",x_test)
np.save("/content/mobel_Data/X_val.npy",x_val)
np.save("/content/mobel_Data/Y_train.npy",y_train)
np.save("/content/mobel_Data/Y_test.npy",y_test)
np.save("/content/mobel_Data/Y_val.npy",y_val)
```
